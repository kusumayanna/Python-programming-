```python
"""


1.What is the difference between a built-in function and a user-defined function? Provide an example of each.

Ans:
- Built-in functions are functions that are provided by Python itself, without needing to define them. 
- Examples include functions like `len()`, `print()`, `sum()`, etc.
- User-defined functions are functions that you create by yourself to perform specific tasks. 
- These are defined using the `def` keyword.

Example of a built-in function:

- Using the built-in function len()
string = "Hello"
print(len(string))  # Output: 5

Example of a user-defined function:**

- Defining a user-defined function to add two numbers
def add_numbers(a, b):
    return a + b

print(add_numbers(3, 5))  # Output: 8


2.How can you pass arguments to a function in Python? Explain the difference between positional arguments and keyword arguments.

Ans:
- Positional arguments are arguments passed to a function in the order they are defined. 
- The first argument passed corresponds to the first parameter in the function definition.
- Keyword arguments are passed to a function using the syntax `parameter=value`. 
- This allows you to specify the arguments by name, irrespective of their position.

Example of positional arguments:

def greet(name, age):
    print(f"Hello {name}, you are {age} years old.")

greet("Alice", 25)  # Output: Hello Alice, you are 25 years old.

Example of keyword arguments:

greet(age=25, name="Alice")  # Output: Hello Alice, you are 25 years old.


3.What is the purpose of the return statement in a function? Can a function have multiple return statements? Explain with an example.

Ans:
- The `return` statement is used to exit a function and optionally pass a value back to the caller. 
- When the function executes a `return`, it stops executing further and sends the result back to the place where the function was called.
- A function can have multiple return statements, often depending on conditions in the function.

Example with multiple return statements:

def check_number(num):
    if num > 0:
        return "Positive"
    elif num < 0:
        return "Negative"
    else:
        return "Zero"

print(check_number(5))   # Output: Positive
print(check_number(-3))  # Output: Negative
print(check_number(0))   # Output: Zero


4.What are lambda functions in Python? How are they different from regular functions? Provide an example where a lambda function can be useful.

Ans:
- Lambda functions are anonymous, small functions defined using the `lambda` keyword. 
- They can have any number of arguments but only one expression. 
- They are typically used for short-term operations and are often passed as arguments to higher-order functions.
- Regular functions are defined using the `def` keyword and can have multiple expressions, more complex logic, and can be reused across different parts of the program.

Example of a lambda function:

- A simple lambda function to square a number
square = lambda x: x ** 2
print(square(4))  # Output: 16


- Use case of lambda function:
Lambda functions are often used with functions like `map()`, `filter()`, and `sorted()`:

numbers = [1, 2, 3, 4, 5]
- Using lambda with map to square each number in the list
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # Output: [1, 4, 9, 16, 25]


5.How does the concept of "scope" apply to functions in Python? Explain the difference between local scope and global scope.

Ans:
- Scope in Python refers to the region in the program where a variable is accessible. 
- There are two main types of scope:
  - Local scope refers to variables that are defined within a function. 
  - These variables are only accessible inside the function.
  - Global scope refers to variables that are defined outside of any function, typically at the top level of a module. 
  - These variables are accessible throughout the module.

Example of local scope:

def my_function():
    local_var = 10  # local variable
    print(local_var)

my_function()  # Output: 10
# print(local_var)  # This will raise an error because local_var is not accessible outside the function.


Example of global scope:

global_var = 20  # global variable

def my_function():
    print(global_var)  # Accessible within the function

my_function()  # Output: 20


6.How can you use the "return" statement in a Python function to return multiple values?

Ans:
In Python, a function can return multiple values as a tuple. 
- simply separate the values with commas, and Python will automatically package them into a tuple.

Example:

def get_coordinates():
    x = 5
    y = 10
    return x, y  # Returning multiple values as a tuple

coords = get_coordinates()
print(coords)  # Output: (5, 10)

# can also unpack the returned tuple into separate variables:
x, y = get_coordinates()
print(x, y)  # Output: 5 10


7.What is the difference between the "pass by value" and "pass by reference" concepts when it comes to function arguments in Python?

Ans:
- Python's argument-passing mechanism is neither strictly pass-by-value nor pass-by-reference but rather a mechanism known as pass-by-object-reference (also called **pass-by-assignment).
  - Immutable objects (e.g., integers, strings, tuples) behave like "pass-by-value" because their value cannot be changed within the function.
  - Mutable objects (e.g., lists, dictionaries) behave like "pass-by-reference" because you can modify their contents within the function.

Example of immutable (pass-by-value-like behavior):

def modify_number(x):
    x = 10  # This doesn't affect the original value
    print(x)

num = 5
modify_number(num)
print(num)  # Output: 5


Example of mutable (pass-by-reference-like behavior):

def modify_list(lst):
    lst.append(4)  # This modifies the original list

my_list = [1, 2, 3]
modify_list(my_list)
print(my_list)  # Output: [1, 2, 3, 4]


8.Create a function that can intake integer or decimal value and do the following operations:
- a. Logarithmic function (log x)
- b. Exponential function (exp(x))
- c. Power function with base 2 (2^x)
- d. Square root

Ans:

import math

def perform_operations(x):
    if x <= 0:
        return "Logarithm not defined for non-positive values"
    
    log_x = math.log(x)  # Logarithmic function
    exp_x = math.exp(x)  # Exponential function
    power_x = 2 ** x  # Power function with base 2
    sqrt_x = math.sqrt(x)  # Square root
    
    return log_x, exp_x, power_x, sqrt_x

result = perform_operations(3)
print(result)  # Output: (1.0986122886681098, 20.085536923187668, 8, 1.7320508075688772)


9.Create a function that takes a full name as an argument and returns the first name and last name.

Ans:

def split_name(full_name):
    names = full_name.split()  # Split the full name into parts
    if len(names) == 2:
        return names[0], names[1]  # Return first and last names
    else:
        return "Please provide a full name with first and last name."

- Example usage
first, last = split_name("John Doe")
print(f"First Name: {first}, Last Name: {last}")  # Output: First Name: John, Last Name: Doe


"""
```
