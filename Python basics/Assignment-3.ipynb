```python
"""
1. Why are functions advantageous to have in your programs?

Ans: 
Reusability: Avoid redundant code by reusing functions.
Modularity: Break down large programs into manageable, logical sections.
Readability: Make code easier to understand with meaningful function names.
Abstraction: Hide complex logic and focus on what a function does.
Easier Debugging and Testing: Isolate and test individual parts of the program.
Encapsulation: Prevent unintended side effects by limiting variable scope.
Maintainability: Simplify code updates by modifying a single function.
Parameterization: Make functions flexible with input arguments.
Separation of Concerns: Organize code by grouping related tasks together.

2. When does the code in a function run: when it's specified or when it's called?

Ans: 
When defined: The function's code is stored in memory, but it does not execute at this point.
When called: The function's code is executed when you call the function by its name, passing any necessary arguments.

3. What statement creates a function?

Ans:
def function_name():
    

4. What is the difference between a function and a function call?

Ans:
Function:
- A function is a block of reusable code that performs a specific task.
- It is defined once using the def keyword (in Python) and can be executed whenever needed by calling it.
- eg: def greet():
    print("Hello!")

Function Call:
- A function call is the action of invoking or executing the function that has been defined.
- When you call a function, the code inside the function is executed
- eg: greet()  



5. How many global scopes are there in a Python program? How many local scopes?

Ans:
Global Scope:
- There is one global scope in a Python program, which exists for the entire program or module.
- The global scope is where top-level variables, functions, and imports are defined.
- eg: variables declared outside any function are part of the global scope.

Local Scope:
- A local scope refers to variables that are defined inside a function or a block of code (like a loop or comprehension).
- Each function call creates its own local scope, and any variables declared inside that function are local to that function.
- There can be multiple local scopes in a program, one for each function call.

6. What happens to variables in a local scope when the function call returns?

Ans:
- When my_function() is called, a local scope is created, and the variable x is defined inside that scope.
- The function executes and prints the value of x.
- After the function finishes executing (when my_function() returns), the local scope is destroyed, and x is no longer accessible.
- If you try to access x outside the function, it results in a NameError, because the variable no longer exists in memory.


7. What is the concept of a return value? Is it possible to have a return value in an expression?

Ans:
- A return value is the value that a function sends back to the part of the program that called it. 
- This value is returned using the return statement, and it can be used for further computations or operations.

Return Value in an Expression:
Yes, it is possible to use a return value in an expression.
- A function can be called and directly use its return value in calculations or other expressions.

8. If a function does not have a return statement, what is the return value of a call to that function?

Ans:
If a function does not have a return statement (or if it reaches the end of the function without executing a return),
it implicitly returns the value **None**.


9. How do you make a function variable refer to the global variable?

Ans:
Use the global keyword inside a function if you want to modify or refer to a variable that exists in the global scope.
Without global, a variable inside the function would be treated as local, and changes wouldn't affect the global variable.
eg: x = 10  

def modify_global():
    global x  
    x = 20  

print(x)  
modify_global()
print(x)  


10. What is the data type of None?

Ans:
The data type of None is NoneType.


11. What does the sentence import areallyourpetsnamederic do?

Ans:
The sentence import areallyourpetsnamederic tries to import a module named areallyourpetsnamederic,
but if this module doesn't exist, Python will raise an ImportError.

12. If you had a bacon() feature in a spam module, what would you call it after importing spam?

Ans:
After importing the spam module, the bacon() function called using spam.bacon().

13. What can you do to save a programme from crashing if it encounters an error?

Ans:
- Use try and except blocks to handle potential errors and avoid crashes.
- Catch specific exceptions to provide more targeted error messages.
- Use else for code that should run if no exceptions occur and finally for cleanup tasks that should always run.

14. What is the purpose of the try clause? What is the purpose of the except clause?

Ans:
Purpose of the try Clause:
- The try clause is used to define a block of code that might raise an exception (an error). 
- It allows you to test code for potential errors while preventing the program from crashing.
- The purpose of the try clause is to attempt running code that might result in an error, without halting the execution of the program.
- If no error occurs in the try block, the program continues normally.
- syntax: 
try:
    # Code that might raise an error
    risky_code()

Purpose of the except Clause:
- The except clause is used to handle exceptions (errors) that occur within the corresponding try block.
- It lets you define how to react when a particular error happens, without crashing the program.
- The purpose of the except clause is to catch and handle errors that occur in the try block, allowing the program to continue running instead of crashing.
- You can specify the type of exception to catch (e.g., ValueError, ZeroDivisionError) or catch any exception.
- syntax:
try:
    # Code that might raise an error
    risky_code()
except SomeException as e:
    # Code to handle the exception
    print(f"An error occurred: {e}")









"""
```
