```python
"""
1. What does an empty dictionary's code look like?

Ans:
An empty dictionary in Python is created using curly braces `{}`.

empty_dict = {}


This creates an empty dictionary with no key-value pairs.


2. What is the value of a dictionary value with the key 'foo' and the value 42?

Ans:
If you have a dictionary like:

my_dict = {'foo': 42}


The value associated with the key `'foo'` is 42.

value = my_dict['foo']  # value will be 42


3. What is the most significant distinction between a dictionary and a list?

Ans:
The most significant distinction between a dictionary and a list is:

- Dictionaries are key-value pairs, where each key is unique and maps to a corresponding value.
- Lists are ordered collections of elements, where elements are indexed by their position (integer index, starting from 0).

Example:
- Dictionary: `{'key1': 'value1', 'key2': 'value2'}` 
- List: `['apple', 'banana', 'cherry']`

4. What happens if you try to access `spam['foo']` if `spam` is `{'bar': 100}`?

Ans:
If you try to access `spam['foo']` when the dictionary is defined as:

spam = {'bar': 100}

You will get a KeyError because `'foo'` is not a key in the dictionary `spam`.

spam = {'bar': 100}
print(spam['foo'])  # This raises a KeyError


To handle this safely, you could use the `get()` method to avoid an error:

value = spam.get('foo', 'default_value')  # Returns 'default_value' if 'foo' is not a key


5. If a dictionary is stored in `spam`, what is the difference between the expressions `'cat' in spam` and `'cat' in spam.keys()`?

Ans:
- `'cat' in spam` checks if 'cat' is a key in the dictionary `spam`.
- `'cat' in spam.keys()`explicitly checks if 'cat' is a key in the dictionary `spam.keys()`, which is a view of all the keys in the dictionary.

Both expressions essentially do the same thing: check if `'cat'` is a key in `spam`. However, `'cat' in spam` is more direct and commonly used.

Example:

spam = {'cat': 'animal', 'dog': 'animal'}

Both return True
print('cat' in spam)         # Output: True
print('cat' in spam.keys())  # Output: True


6. If a dictionary is stored in `spam`, what is the difference between the expressions `'cat' in spam` and `'cat' in spam.values()`?

Ans:
- `'cat' in spam` checks if 'cat' is a key in the dictionary `spam`.
- `'cat' in spam.values()` checks if 'cat' is a value in the dictionary `spam`.

So, the difference is that one checks for keys and the other checks for values.

Example:

spam = {'animal': 'cat', 'bird': 'parrot'}

print('cat' in spam)          # Output: False (since 'cat' is not a key)
print('cat' in spam.values()) # Output: True  (since 'cat' is a value)


7. What is a shortcut for the following code?

Ans:
if 'color' not in spam:
    spam['color'] = 'black'


You can use the `setdefault()` method as a shortcut. It checks if the key exists; if not, it sets the default value.

spam.setdefault('color', 'black')

This will set `'color'` to `'black'` only if `'color'` does not already exist in the dictionary.


8. How do you "pretty print" dictionary values using which module and function?

Ans:
To pretty print dictionary values, you can use the `pprint` module and its `pprint()` function.

Example:

import pprint

my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}

Pretty print the dictionary
pprint.pprint(my_dict)

This will display the dictionary in a nicely formatted, readable way, especially if the dictionary is large or nested.


"""
```
